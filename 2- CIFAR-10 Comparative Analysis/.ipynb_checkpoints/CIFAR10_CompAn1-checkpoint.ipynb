```python
import numpy as np
import pandas as pd
import os
import pickle
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report, confusion_matrix

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.models import Sequential, clone_model
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.utils import to_categorical

from tensorflow.keras.layers import BatchNormalization
from tensorflow.keras.callbacks import ReduceLROnPlateau, EarlyStopping
from tensorflow.keras import regularizers

from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
cifar10_dir = r"D:\Data_and_AI\Datasets\cifar10\cifar-10-batches-py" 
```

```python
def unpickle(file):
    """Load CIFAR-10 batch file."""
    with open(file, 'rb') as fo:
        data_dict = pickle.load(fo, encoding='bytes')  # Load with byte encoding
    return data_dict
```

```python
x_train = []
y_train = []

for i in range(1, 6):  #to get through the 5 training batches
    batch = unpickle(os.path.join(cifar10_dir, f"data_batch_{i}"))
    x_train.append(batch[b'data'])   # Image data
    y_train.append(batch[b'labels']) # Labels

x_train = np.concatenate(x_train, axis=0)  # Stack all batches
y_train = np.concatenate(y_train, axis=0)  # Stack labels

# Load test batch
test_batch = unpickle(os.path.join(cifar10_dir, "test_batch"))
x_test = test_batch[b'data']
y_test = np.array(test_batch[b'labels'])

# CIFAR-10 has 10 classes
label_names = unpickle(os.path.join(cifar10_dir, "batches.meta"))[b'label_names']
label_names = [label.decode('utf-8') for label in label_names]  # Convert byte strings to normal strings

# Reshape image data to (num_images, 32, 32, 3)
x_train = x_train.reshape(-1, 3, 32, 32).transpose(0, 2, 3, 1)  # Convert from (N, 3, 32, 32) to (N, 32, 32, 3)
x_test = x_test.reshape(-1, 3, 32, 32).transpose(0, 2, 3, 1)

# Normalize pixel values to [0, 1]
x_train = x_train / 255.0
x_test = x_test / 255.0

print(f"Training Set: {x_train.shape}, Test Set: {x_test.shape}")
print(f"Classes: {label_names}")
```

```python
# Split the original training set (50,000 images) into Train (80%) and Validation (20%)
x_train, x_val, y_train, y_val = train_test_split(x_train, y_train, test_size=0.2, random_state=42, stratify=y_train)

# Print dataset shapes
print(f"New Training Set: {x_train.shape}, Labels: {y_train.shape}")
print(f"Validation Set: {x_val.shape}, Labels: {y_val.shape}")
print(f"Test Set (unchanged): {x_test.shape}, Labels: {y_test.shape}")
```

```python
def show_images(images, labels, title, num_images=5):
    plt.figure(figsize=(10, 2))  # Set figure size
    for i in range(num_images):
        plt.subplot(1, num_images, i + 1)
        index = np.random.randint(0, len(images))  # Pick a random index
        plt.imshow(images[index])  
        plt.title(label_names[labels[index]])  
        plt.axis("off")  # Hide axes
    plt.show()

# Show images from each dataset
show_images(x_train, y_train, "Training Set")
show_images(x_val, y_val, "Validation Set")
show_images(x_test, y_test, "Test Set")
```

## Models

### Model 1

```python
# One-hot encode labels for training, validation, and test sets
y_train_cat = to_categorical(y_train, num_classes=10)
y_val_cat = to_categorical(y_val, num_classes=10)
y_test_cat = to_categorical(y_test, num_classes=10)
```

```python
model = Sequential([
    keras.Input(shape=(32, 32, 3)),
    layers.Conv2D(32, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dense(10, activation='softmax')
])
```

```python
model.compile(optimizer=Adam(),
              loss='categorical_crossentropy',
              metrics=['accuracy'])
model.summary()
```

```python
history = model.fit(
    x_train, y_train_cat,
    epochs=10,
    batch_size=64,
    validation_data=(x_val, y_val_cat)
)
```

```python
plt.figure(figsize=(12, 5))

# Plot Accuracy
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Accuracy over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

# Plot Loss
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Loss over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.show()
```

```python
test_loss, test_acc = model.evaluate(x_test, y_test_cat)
print(f"Test Accuracy: {test_acc:.4f}")
```

```python
# Generate predictions for the test set
y_pred_probs = model.predict(x_test)
y_pred_classes = np.argmax(y_pred_probs, axis=1)

# Classification Report
print(classification_report(y_test, y_pred_classes, target_names=label_names))

# Confusion Matrix
conf_mat = sns.heatmap(
    pd.DataFrame(confusion_matrix(y_test, y_pred_classes), 
                 index=label_names, columns=label_names),
    annot=True, fmt='d', cmap='Blues'
)
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()
```

### Model 3:
#### A bigger model (with learning rate scheduling)

```python
model_3 = Sequential([
    keras.Input(shape=(32, 32, 3)),

    layers.Conv2D(32, (3, 3), activation='relu', padding='same'),
    layers.Conv2D(32, (3, 3), activation='relu', padding='same'),
    layers.MaxPooling2D((2, 2)),
    layers.Dropout(0.25),

    layers.Conv2D(64, (3, 3), activation='relu', padding='same'),
    layers.Conv2D(64, (3, 3), activation='relu', padding='same'),
    layers.MaxPooling2D((2, 2)),
    layers.Dropout(0.25),

    layers.Conv2D(128, (3, 3), activation='relu', padding='same'),
    layers.Conv2D(128, (3, 3), activation='relu', padding='same'),
    layers.MaxPooling2D((2, 2)),
    layers.Dropout(0.25),

    layers.Conv2D(256, (3, 3), activation='relu', padding='same'),
    layers.Conv2D(256, (3, 3), activation='relu', padding='same'),
    layers.MaxPooling2D((2, 2)),
    layers.Dropout(0.25),

    layers.Flatten(),
    layers.Dense(256, activation='relu'),
    layers.Dropout(0.5),
    layers.Dense(10, activation='softmax')
])
```

```python
model_3.compile(optimizer=Adam(learning_rate=0.0005),
              loss='categorical_crossentropy',
              metrics=['accuracy'])

print(model_3.summary())
```

```python
# Implement Learning Rate Scheduling
lr_scheduler = ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=3, min_lr=0.0001)
```

```python
# Train model_3 with learning rate scheduling
history_3_lr = model_3.fit(
    x_train, y_train_cat,
    epochs=30, # Let's train for a bit longer to see the effect of scheduling
    batch_size=64,
    validation_data=(x_val, y_val_cat),
    callbacks=[lr_scheduler]
)
```

```python
# Evaluate the model
plt.figure(figsize=(12, 5))

# Plot Accuracy
plt.subplot(1, 2, 1)
plt.plot(history_3_lr.history['accuracy'], label='Train Accuracy (Model 3 with LR)')
plt.plot(history_3_lr.history['val_accuracy'], label='Validation Accuracy (Model 3 with LR)')
plt.title('Accuracy over Epochs (Model 3 with LR)')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

# Plot Loss
plt.subplot(1, 2, 2)
plt.plot(history_3_lr.history['loss'], label='Train Loss (Model 3 with LR)')
plt.plot(history_3_lr.history['val_loss'], label='Validation Loss (Model 3 with LR)')
plt.title('Loss over Epochs (Model 3 with LR)')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.show()
```

```python
test_loss_3_lr, test_acc_3_lr = model_3.evaluate(x_test, y_test_cat)
print(f"Test Accuracy (Model 3 with LR): {test_acc_3_lr:.4f}")
```

```python
# Generate predictions for the test set
y_pred_probs_3_lr = model_3.predict(x_test)
y_pred_classes_3_lr = np.argmax(y_pred_probs_3_lr, axis=1)

# Classification Report
print("Classification Report (Model 3 with LR):")
print(classification_report(y_test, y_pred_classes_3_lr, target_names=label_names))

# Confusion Matrix
conf_mat_3_lr = sns.heatmap(
    pd.DataFrame(confusion_matrix(y_test, y_pred_classes_3_lr),
                 index=label_names, columns=label_names),
    annot=True, fmt='d', cmap='Blues'
)
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix (Model 3 with LR)")
plt.show()
```

### Model 4:
#### Same Architecture, added L2, EarlyStopping

```python
model_4 = Sequential([
    keras.Input(shape=(32, 32, 3)),

    layers.Conv2D(32, (3, 3), activation='relu', padding='same', kernel_regularizer=regularizers.l2(0.0001)),
    layers.Conv2D(32, (3, 3), activation='relu', padding='same', kernel_regularizer=regularizers.l2(0.0001)),
    layers.MaxPooling2D((2, 2)),
    layers.Dropout(0.25),

    layers.Conv2D(64, (3, 3), activation='relu', padding='same', kernel_regularizer=regularizers.l2(0.0001)),
    layers.Conv2D(64, (3, 3), activation='relu', padding='same', kernel_regularizer=regularizers.l2(0.0001)),
    layers.MaxPooling2D((2, 2)),
    layers.Dropout(0.25),

    layers.Conv2D(128, (3, 3), activation='relu', padding='same', kernel_regularizer=regularizers.l2(0.0001)),
    layers.Conv2D(128, (3, 3), activation='relu', padding='same', kernel_regularizer=regularizers.l2(0.0001)),
    layers.MaxPooling2D((2, 2)),
    layers.Dropout(0.25),

    layers.Conv2D(256, (3, 3), activation='relu', padding='same', kernel_regularizer=regularizers.l2(0.0001)),
    layers.Conv2D(256, (3, 3), activation='relu', padding='same', kernel_regularizer=regularizers.l2(0.0001)),
    layers.MaxPooling2D((2, 2)),
    layers.Dropout(0.25),

    layers.Flatten(),
    layers.Dense(256, activation='relu', kernel_regularizer=regularizers.l2(0.0001)),
    layers.Dropout(0.5),
    layers.Dense(10, activation='softmax', kernel_regularizer=regularizers.l2(0.0001))
])
```

```python
model_4.compile(optimizer=Adam(learning_rate=0.0005),
              loss='categorical_crossentropy',
              metrics=['accuracy'])

print(model_4.summary())
```

```python
# Implement Learning Rate Scheduling
lr_scheduler = ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=3, min_lr=0.0001)

# Implement Early Stopping
early_stopping = EarlyStopping(monitor='val_accuracy', patience=3, restore_best_weights=True)
```

```python
# Train model_4 with L2 regularization, dropout, LR scheduling, and early stopping
history_4 = model_4.fit(
    x_train, y_train_cat,
    epochs=30,
    batch_size=64,
    validation_data=(x_val, y_val_cat),
    callbacks=[lr_scheduler, early_stopping]
)
```

```python
# Evaluate the model
plt.figure(figsize=(12, 5))

# Plot Accuracy
plt.subplot(1, 2, 1)
plt.plot(history_4.history['accuracy'], label='Train Accuracy (Model 4)')
plt.plot(history_4.history['val_accuracy'], label='Validation Accuracy (Model 4)')
plt.title('Accuracy over Epochs (Model 4)')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

# Plot Loss
plt.subplot(1, 2, 2)
plt.plot(history_4.history['loss'], label='Train Loss (Model 4)')
plt.plot(history_4.history['val_loss'], label='Validation Loss (Model 4)')
plt.title('Loss over Epochs (Model 4)')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.show()
```

```python
test_loss_4, test_acc_4 = model_4.evaluate(x_test, y_test_cat)
print(f"Test Accuracy (Model 4): {test_acc_4:.4f}")
```

```python
# Generate predictions for the test set
y_pred_probs_4 = model_4.predict(x_test)
y_pred_classes_4 = np.argmax(y_pred_probs_4, axis=1)

# Classification Report
print("Classification Report (Model 4):")
print(classification_report(y_test, y_pred_classes_4, target_names=label_names))

# Confusion Matrix
conf_mat_4 = sns.heatmap(
    pd.DataFrame(confusion_matrix(y_test, y_pred_classes_4),
                 index=label_names, columns=label_names),
    annot=True, fmt='d', cmap='Blues'
)
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix (Model 4)")
plt.show()
```

### Model 5:
#### Same as Model 4 but with Batch Normalization added

```python
# Define model_5 architecture (based on model_4 with added BatchNormalization)
model_5 = Sequential([
    keras.Input(shape=(32, 32, 3)),

    layers.Conv2D(32, (3, 3), padding='same', kernel_regularizer=regularizers.l2(0.0001)),
    BatchNormalization(),
    layers.Activation('relu'),
    layers.Conv2D(32, (3, 3), padding='same', kernel_regularizer=regularizers.l2(0.0001)),
    BatchNormalization(),
    layers.Activation('relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Dropout(0.25),

    layers.Conv2D(64, (3, 3), padding='same', kernel_regularizer=regularizers.l2(0.0001)),
    BatchNormalization(),
    layers.Activation('relu'),
    layers.Conv2D(64, (3, 3), padding='same', kernel_regularizer=regularizers.l2(0.0001)),
    BatchNormalization(),
    layers.Activation('relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Dropout(0.25),

    layers.Conv2D(128, (3, 3), padding='same', kernel_regularizer=regularizers.l2(0.0001)),
    BatchNormalization(),
    layers.Activation('relu'),
    layers.Conv2D(128, (3, 3), padding='same', kernel_regularizer=regularizers.l2(0.0001)),
    BatchNormalization(),
    layers.Activation('relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Dropout(0.25),

    layers.Conv2D(256, (3, 3), padding='same', kernel_regularizer=regularizers.l2(0.0001)),
    BatchNormalization(),
    layers.Activation('relu'),
    layers.Conv2D(256, (3, 3), padding='same', kernel_regularizer=regularizers.l2(0.0001)),
    BatchNormalization(),
    layers.Activation('relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Dropout(0.25),

    layers.Flatten(),
    layers.Dense(256, kernel_regularizer=regularizers.l2(0.0001)),
    BatchNormalization(),
    layers.Activation('relu'),
    layers.Dropout(0.5),
    layers.Dense(10, activation='softmax', kernel_regularizer=regularizers.l2(0.0001))
])
```

```python
# Compile model_5
model_5.compile(optimizer=Adam(learning_rate= 0.0005),
              loss='categorical_crossentropy',
              metrics=['accuracy'])

print(model_5.summary())
```

```python
# Implement Learning Rate Scheduling
lr_scheduler = ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=3, min_lr=0.0001)

# Implement Early Stopping with patience of 3 (you can adjust this)
early_stopping = EarlyStopping(monitor='val_accuracy', patience=5, restore_best_weights=True)
```

```python
# Train model_5
history_5 = model_5.fit(
    x_train, y_train_cat,
    epochs=50, # You can adjust the number of epochs
    batch_size=64,
    validation_data=(x_val, y_val_cat),
    callbacks=[lr_scheduler, early_stopping]
)
```

```python
# Evaluate the model
plt.figure(figsize=(12, 5))

# Plot Accuracy
plt.subplot(1, 2, 1)
plt.plot(history_5.history['accuracy'], label='Train Accuracy (Model 5)')
plt.plot(history_5.history['val_accuracy'], label='Validation Accuracy (Model 5)')
plt.title('Accuracy over Epochs (Model 5)')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

# Plot Loss
plt.subplot(1, 2, 2)
plt.plot(history_5.history['loss'], label='Train Loss (Model 5)')
plt.plot(history_5.history['val_loss'], label='Validation Loss (Model 5)')
plt.title('Loss over Epochs (Model 5)')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.show()
```

```python
num_samples_to_show = 3
sample_indices = np.random.choice(len(x_train), num_samples_to_show, replace=False)
sample_images = x_train[sample_indices]
sample_labels = y_train[sample_indices]
label_names_array = np.array(label_names)

# Create ImageDataGenerator with multiple augmentations
datagen_augmented = ImageDataGenerator(
    horizontal_flip=True,
    width_shift_range=0.1,
    height_shift_range=0.1,
    rotation_range=15
)

fig, axes = plt.subplots(nrows=num_samples_to_show, ncols=2, figsize=(8, 5 * num_samples_to_show))

for i in range(num_samples_to_show):
    original_img = sample_images[i]
    axes[i, 0].imshow(original_img)
    axes[i, 0].set_title(f"Original: {label_names_array[sample_labels[i]]}")
    axes[i, 0].axis('off')

    # Reshape to (1, height, width, channels) for the generator
    img = np.expand_dims(original_img, axis=0)
    augmented_images = datagen_augmented.flow(img, batch_size=1, shuffle=False)

    # Get the single augmented image
    augmented_img = next(augmented_images)[0]
    augmented_img_uint8 = (augmented_img * 255).astype('uint8')
    axes[i, 1].imshow(augmented_img_uint8)
    axes[i, 1].set_title(f"Augmented")
    axes[i, 1].axis('off')

plt.tight_layout()
plt.show()
```

### Model 5 (with Data Augmentation)

```python
#create a copy of model_5
model_5_aug = Sequential([
    keras.Input(shape=(32, 32, 3)),

    layers.Conv2D(32, (3, 3), padding='same', kernel_regularizer=regularizers.l2(0.0001)),
    BatchNormalization(),
    layers.Activation('relu'),
    layers.Conv2D(32, (3, 3), padding='same', kernel_regularizer=regularizers.l2(0.0001)),
    BatchNormalization(),
    layers.Activation('relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Dropout(0.25),

    layers.Conv2D(64, (3, 3), padding='same', kernel_regularizer=regularizers.l2(0.0001)),
    BatchNormalization(),
    layers.Activation('relu'),
    layers.Conv2D(64, (3, 3), padding='same', kernel_regularizer=regularizers.l2(0.0001)),
    BatchNormalization(),
    layers.Activation('relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Dropout(0.25),

    layers.Conv2D(128, (3, 3), padding='same', kernel_regularizer=regularizers.l2(0.0001)),
    BatchNormalization(),
    layers.Activation('relu'),
    layers.Conv2D(128, (3, 3), padding='same', kernel_regularizer=regularizers.l2(0.0001)),
    BatchNormalization(),
    layers.Activation('relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Dropout(0.25),

    layers.Conv2D(256, (3, 3), padding='same', kernel_regularizer=regularizers.l2(0.0001)),
    BatchNormalization(),
    layers.Activation('relu'),
    layers.Conv2D(256, (3, 3), padding='same', kernel_regularizer=regularizers.l2(0.0001)),
    BatchNormalization(),
    layers.Activation('relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Dropout(0.25),

    layers.Flatten(),
    layers.Dense(256, kernel_regularizer=regularizers.l2(0.0001)),
    BatchNormalization(),
    layers.Activation('relu'),
    layers.Dropout(0.5),
    layers.Dense(10, activation='softmax', kernel_regularizer=regularizers.l2(0.0001))
])
```

```python
model_5_aug.compile(optimizer=Adam(learning_rate= 0.0005),
              loss='categorical_crossentropy',
              metrics=['accuracy'])
print(model_5_aug.summary())
```

```python
# Create the training data generator
train_generator = datagen_augmented.flow(
    x_train,
    y_train_cat, 
    batch_size=64,
    shuffle=True
)
```

```python
# Implement Learning Rate Scheduling
lr_scheduler = ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=3, min_lr=0.0001)

# Implement Early Stopping with patience of 3 (you can adjust this)
early_stopping = EarlyStopping(monitor='val_accuracy', patience=5, restore_best_weights=True)
```

```python
history_model_5_aug = model_5_aug.fit(
    train_generator,
    epochs=50, # You can adjust the number of epochs
    validation_data=(x_val, y_val_cat),
    callbacks=[lr_scheduler, early_stopping]
)
```
