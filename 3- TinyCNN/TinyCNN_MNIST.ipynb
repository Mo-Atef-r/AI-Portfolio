```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from tensorflow.keras.datasets import mnist
from sklearn.model_selection import train_test_split
from tensorflow.keras.utils import to_categorical

from tensorflow.keras.models import Sequential
from tensorflow.keras import layers
from tensorflow.keras import regularizers
from tensorflow.keras.callbacks import EarlyStopping

from sklearn.metrics import classification_report, confusion_matrix
```

```python
(x_train_original, y_train), (x_test, y_test) = mnist.load_data()
validation_size = int(x_test.shape[0])
x_train, x_val, y_train, y_val = train_test_split(
    x_train_original, y_train, test_size=validation_size, random_state=42, stratify=y_train)


print("\nShape of x_train:", x_train.shape)
print("Shape of x_val:", x_val.shape)
print("Shape of x_test:", x_test.shape)
print("Shape of y_train:", y_train.shape)
print("Sample of y_train:", y_train[:5])
```

```python
# 5. Perform one-hot encoding 
num_classes = 10 
y_train_cat = to_categorical(y_train, num_classes)
y_val_cat = to_categorical(y_val, num_classes)
y_test_cat = to_categorical(y_test, num_classes)
print("Shape of one-hot encoded y_train_cat:", y_train_cat.shape)
print("Sample of one-hot encoded y_train_cat:\n", y_train_cat[:2])
```

```python
num_samples_to_show = 10
indices = np.random.choice(len(x_train), num_samples_to_show, replace=False)

plt.figure(figsize=(12, 5))
for i, index in enumerate(indices):
    plt.subplot(2, 5, i + 1)
    plt.imshow(x_train[index], cmap='gray')
    plt.title(f"Label: {y_train[index]}")
    plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
x_train = x_train.astype('float32') / 255.0
x_val = x_val.astype('float32') / 255.0
x_test = x_test.astype('float32') / 255.0

x_train = x_train.reshape(-1, 28, 28, 1)
x_val = x_val.reshape(-1, 28, 28, 1)
x_test = x_test.reshape(-1, 28, 28, 1)
```

## Models

### Model 1: 225K Parameters

```python
l2_rate = 0.001
```

```python
model = Sequential ([
    layers.Input((28,28,1)),
    
    layers.Conv2D(32,(3,3), activation= 'relu', kernel_regularizer=regularizers.l2(l2_rate)),
    layers.MaxPooling2D((2,2)),

    layers.Conv2D(64,(3,3), activation= 'relu', kernel_regularizer=regularizers.l2(l2_rate)),
    layers.MaxPooling2D((2,2)),
    
    layers.Flatten(),

    layers.Dense(128, activation= 'relu', kernel_regularizer=regularizers.l2(l2_rate)),
    layers.Dense(10, activation= 'softmax')
])
```

```python
model.compile(optimizer = 'adam', loss='categorical_crossentropy',
              metrics=['accuracy'])
print(model.summary())
```

```python
epochs = 20 
early_stopping = EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)
```

```python
history = model.fit(x_train, y_train_cat,
                    epochs=epochs,
                    validation_data=(x_val, y_val_cat),
                    callbacks=[early_stopping])
```

```python
plt.figure(figsize=(10, 5))
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.title('Training and Validation Accuracy')
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 5))
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Training and Validation Loss')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Evaluate the model on the test set
_, accuracy = model.evaluate(x_test, y_test_cat, verbose=0)
print(f'Test Accuracy: {accuracy:.4f}')
```

```python
# Make predictions on the test set
y_pred_probabilities = model.predict(x_test)
y_pred_classes = np.argmax(y_pred_probabilities, axis=1)

# Convert one-hot encoded test labels back to class labels
y_true_classes = np.argmax(y_test_cat, axis=1)
```

```python
class_names = [str(i) for i in range(10)]  # Class names for digits 0-9
report = classification_report(y_true_classes, y_pred_classes, target_names=class_names)
print("\nClassification Report:")
print(report)
```

```python
cm = confusion_matrix(y_true_classes, y_pred_classes)

plt.figure(figsize=(10, 8))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=class_names, yticklabels=class_names)
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Confusion Matrix')
plt.show()
```
