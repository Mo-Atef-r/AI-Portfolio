```python
import os
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

from concurrent.futures import ThreadPoolExecutor
from PIL import Image

from skimage.feature import local_binary_pattern, hog

from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

from sklearn.ensemble import IsolationForest
from sklearn.metrics import classification_report, precision_recall_fscore_support, roc_auc_score, RocCurveDisplay

import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, UpSampling2D, Dropout, Dense, concatenate
from tensorflow.keras.optimizers import Adam

from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
from tensorflow.keras.regularizers import l2
```

```python
BASE_PATH = r"D:\Data_and_AI\Datasets\Microscopic Images of Integrated Circuits\Files\Anomaly_train"
IMAGE_SIZE = (128, 128)
BATCH_SIZE = 32
```

```python
def load_images(category):
    """Load all images for a category with parallel processing"""
    files = sorted([f for f in os.listdir(BASE_PATH) if f.startswith(category)])
    images = []
    
    def load_image(filename):
        img = Image.open(os.path.join(BASE_PATH, filename)).convert('L')
        img = img.resize((128, 128))  # ONLY ADDED THIS LINE
        return np.array(img)
    
    with ThreadPoolExecutor() as executor:
        images = list(executor.map(load_image, files))
    
    return {
        'all': np.array(images),  # Full dataset as NumPy array
        'samples': images[:NUM_SAMPLES]  # Samples for visualization
    }
```

```python
print("Loading datasets...")
datasets = {
    'train_normal': load_images('train_normal'),
    'test_normal': load_images('test_normal'),
    'test_abnormal': load_images('test_abnormal')
}
print("DATASETS LOADED SUCCESSFULLY")
```

```python
# Display samples
def show_samples(samples, title):
    plt.figure(figsize=(15, 5))
    for i, img in enumerate(samples, 1):
        plt.subplot(1, NUM_SAMPLES, i)
        plt.imshow(img, cmap='gray', vmin=0, vmax=255)
        plt.title(f"{title}\nShape: {img.shape}")
        plt.axis('off')
    plt.tight_layout()
    plt.show()

print("\nVisualizing samples...")
#show_samples(datasets['train_normal']['samples'], "Train Normal")
show_samples(datasets['test_normal']['samples'], "Test Normal")
show_samples(datasets['test_abnormal']['samples'], "Test Abnormal")
```

```python
# Full datasets converted to NumPy arrays
X_train = datasets['train_normal']['all']
y_train = np.zeros(len(X_train))  # 0 = normal

X_test_normal = datasets['test_normal']['all']
X_test_abnormal = datasets['test_abnormal']['all']

X_test = np.concatenate([X_test_normal, X_test_abnormal])
y_test = np.concatenate([
    np.zeros(len(X_test_normal)),  # 0 = normal
    np.ones(len(X_test_abnormal))  # 1 = abnormal
])

print("\nFinal Dataset Shapes:")
print(f"X_train: {X_train.shape} (dtype: {X_train.dtype})")
print(f"X_test: {X_test.shape} (dtype: {X_test.dtype})")
print(f"y_train: {y_train.shape} (dtype: {y_train.dtype})")
print(f"y_test: {y_test.shape} (dtype: {y_test.dtype})")

# Normalize data
X_train = X_train.astype('float32') / 255.
X_test = X_test.astype('float32') / 255.

# Reshape for CNN
X_train = X_train.reshape(X_train.shape[0], 128, 128, 1)
X_test = X_test.reshape(X_test.shape[0], 128, 128, 1)
```

```python
input_img = Input(shape=(128, 128, 1))

# Encoder
x = Conv2D(16, (3, 3), activation='relu', padding='same', kernel_regularizer=l2(1e-5))(input_img)
x = MaxPooling2D((2, 2), padding='same')(x)
x = Dropout(0.1)(x)  # 128 -> 64

x = Conv2D(32, (3, 3), activation='relu', padding='same', kernel_regularizer=l2(1e-5))(x)
x = MaxPooling2D((2, 2), padding='same')(x)
x = Dropout(0.2)(x)  # 64 -> 32

x = Conv2D(64, (3, 3), activation='relu', padding='same', kernel_regularizer=l2(1e-5))(x)
encoded = MaxPooling2D((2, 2), padding='same')(x)  # 32 -> 16

# Decoder
x = Conv2D(64, (3, 3), activation='relu', padding='same', kernel_regularizer=l2(1e-5))(encoded)
x = UpSampling2D((2, 2))(x)  # 16 -> 32
x = Dropout(0.2)(x)

x = Conv2D(32, (3, 3), activation='relu', padding='same', kernel_regularizer=l2(1e-5))(x)
x = UpSampling2D((2, 2))(x)  # 32 -> 64
x = Dropout(0.1)(x)

x = Conv2D(16, (3, 3), activation='relu', padding='same', kernel_regularizer=l2(1e-5))(x)
x = UpSampling2D((2, 2))(x)  # 64 -> 128

decoded = Conv2D(1, (3, 3), activation='sigmoid', padding='same')(x)

autoencoder = Model(input_img, decoded)
```

```python
autoencoder.compile(optimizer='adam', loss='mse')
print(autoencoder.summary())
```

```python
# Setup callbacks for early stopping and model checkpointing
early_stopping = EarlyStopping(
    monitor='val_loss',
    patience=3,  
    verbose=1,
    restore_best_weights=True
)

model_checkpoint = ModelCheckpoint(
    'best_autoencoder.keras',
    monitor='val_loss',
    verbose=1,
    save_best_only=True
)
```

```python
# Train model (on normal samples only)
history = autoencoder.fit(
    X_train, X_train,
    epochs=100,  # Set higher, early stopping will prevent overfitting
    batch_size=64,  # Smaller batch size
    shuffle=True,
    validation_split=0.15,
    callbacks=[early_stopping, model_checkpoint],
    verbose=1
)
```

```python
# Plot training history
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='upper right')
```

```python
# Calculate reconstruction error on test set
reconstructions = autoencoder.predict(X_test)
mse = np.mean(np.square(X_test - reconstructions), axis=(1,2,3))

# Find optimal threshold using validation set (create a validation set from training)
val_size = int(0.2 * len(X_train))
X_val = X_train[-val_size:]
reconstructions_val = autoencoder.predict(X_val)
mse_val = np.mean(np.square(X_val - reconstructions_val), axis=(1,2,3))
```

```python
# Create artificial anomalies in validation set for threshold tuning
# Randomly select 10% of validation samples to treat as "anomalies"
np.random.seed(42)
val_anomaly_idx = np.random.choice(val_size, size=int(0.1*val_size), replace=False)
y_val = np.zeros(val_size)
y_val[val_anomaly_idx] = 1

# Add noise to these "anomalies" to simulate defects
for idx in val_anomaly_idx:
    noise = np.random.normal(0, 0.5, X_val[idx].shape)
    X_val[idx] = np.clip(X_val[idx] + noise, 0, 1)

# Recalculate MSE for artificially modified validation set
reconstructions_val = autoencoder.predict(X_val)
mse_val = np.mean(np.square(X_val - reconstructions_val), axis=(1,2,3))

# Find optimal threshold
thresholds = np.linspace(np.percentile(mse_val, 80), np.percentile(mse_val, 99), 100)
f1_scores = []

for threshold in thresholds:
    y_pred = (mse_val > threshold).astype(int)
    _, _, f1, _ = precision_recall_fscore_support(y_val, y_pred, average='binary', zero_division=0)
    f1_scores.append(f1)

optimal_threshold = thresholds[np.argmax(f1_scores)]
print(f"Optimal threshold: {optimal_threshold:.6f}")
```

```python
# Apply to test set
y_pred = (mse > optimal_threshold).astype(int)

# Evaluate performance
precision, recall, f1, _ = precision_recall_fscore_support(y_test, y_pred, average='binary')
auc = roc_auc_score(y_test, mse)

print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1 Score: {f1:.4f}")
print(f"AUC-ROC: {auc:.4f}")
```

```python
# Plot MSE distributions to visualize separation
plt.subplot(1, 2, 2)
plt.hist(mse[y_test==0], bins=50, alpha=0.5, label='Normal')
plt.hist(mse[y_test==1], bins=50, alpha=0.5, label='Anomaly')
plt.axvline(optimal_threshold, color='r', linestyle='--', label='Threshold')
plt.title('MSE Distribution')
plt.xlabel('Reconstruction Error (MSE)')
plt.ylabel('Frequency')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Visualize some examples
def plot_results(original, reconstructed, mse_val, is_anomaly, pred_anomaly):
    plt.figure(figsize=(20, 4))
    
    plt.subplot(1, 4, 1)
    plt.imshow(original.reshape(128, 128), cmap='gray')
    plt.title("Original")
    plt.axis('off')
    
    plt.subplot(1, 4, 2)
    plt.imshow(reconstructed.reshape(128, 128), cmap='gray')
    plt.title("Reconstructed")
    plt.axis('off')
    
    plt.subplot(1, 4, 3)
    diff = np.abs(original - reconstructed).reshape(128, 128)
    plt.imshow(diff, cmap='hot')
    plt.title("Difference Map")
    plt.colorbar()
    plt.axis('off')
    
    plt.subplot(1, 4, 4)
    plt.text(0.5, 0.5, f"MSE: {mse_val:.6f}\nTrue: {'Anomaly' if is_anomaly else 'Normal'}\nPred: {'Anomaly' if pred_anomaly else 'Normal'}", 
             horizontalalignment='center', verticalalignment='center', fontsize=14)
    plt.axis('off')
    
    plt.tight_layout()
    plt.show()
```

```python
# Plot some normal and anomaly examples
normal_idx = np.where(y_test == 0)[0][:3]
anomaly_idx = np.where(y_test == 1)[0][:3]

for idx in normal_idx:
    plot_results(X_test[idx], reconstructions[idx], mse[idx], 
                 is_anomaly=(y_test[idx]==1), pred_anomaly=(y_pred[idx]==1))

for idx in anomaly_idx:
    plot_results(X_test[idx], reconstructions[idx], mse[idx], 
                 is_anomaly=(y_test[idx]==1), pred_anomaly=(y_pred[idx]==1))
```
